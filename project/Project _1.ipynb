### Parent class : holds user account info and show the user details 


### Child class : about bank and holds the balance and can  can make deposits , withdraw and show the final balance  functions  .....


```python
#frist we create the perant class
class big_bank:
#the constractor of the class
    def __init__(self,fristName,lastName,age,phoneNumber,address,nationalId,job) :
        self.fristName=fristName
        self.lastName=lastName
        self.age=age
        self.phoneNumber=phoneNumber
        self.address=address
        self.nationalId=nationalId
        self.job=job

# method to display all user info
    def display_Info(self):
        info=[]
        info.append(self.fristName)
        info.append(self.lastName)
        info.append(self.age)
        info.append(self.phoneNumber)
        info.append(self.nationalId)
        info.append(self.address)
        info.append(self.job)
        return info


      


# customer1 = big_bank('mahmoud', 'ahmed',22,1012021,'st.213',123123123,'doctor')
# customer1.display_Info()

        

```

```python

#the child object enharite the perant child
class child_bank(big_bank):
    #the constractor of the class
    def __init__(self,fristName,lastName,age,phoneNumber,address,nationalId,job) :
        #give the parent his attributes
        super().__init__(fristName,lastName,age,phoneNumber,address,nationalId,job)
        #min balance shoud be 1000
        self.balance=1000

    
    #funcation to display the balance
    def display_balance(self):
        print('your balance is {}'.format(self.balance))
        print('-----------------------------------')
    #the deposite funcation and it call the display_balance
    def deposit(self):
        new_deposite=int(input('please enter the new deposite'))
        self.balance+=new_deposite
        print('deposite done')
    #the withdraw function and it call the display_balance
    def withdraw(self):
        new_withdraw=int(input('please enter the new withdraw'))
        if(new_withdraw>self.balance):
            print('you cann\'t withdraw more than your balance')
        else:
            self.balance-=new_withdraw
            print('withdraw done')

    
```

```python
print("Welcome to your Bank")
print('----------------------------------')
#taking the input from the user 
Fristname = input("Enter FristName ")
LastName = input("Enter LastName ")
age = input("Enter age ")
nationalId = input("Enter national id ")
phone_num = input("Enter phone number ")
job=input('enter job')
address=input('enter address')

user1 = child_bank(Fristname,LastName,age,phone_num,address,nationalId,job)

# the bank system 
while True:

    answer = input("Enter d , w , b or I for get deposite , withdraw , balance or information in the same order")
    if answer == 'd' or answer=='D':
        user1.deposit()
        user1.display_balance()
        
    elif answer == 'w' or answer=='W':
        user1.withdraw()
        user1.display_balance()

    elif answer == 'b' or answer=='B':
       user1.display_balance()
    elif answer == 'i' or answer=='I':
       print(user1.display_Info())
       print('-------------------------------------')
        
    continue_The_operations= input("Do you want another operation?(y or n)")
    if continue_The_operations =='y' or continue_The_operations =='Y':
        continue
    elif continue_The_operations == 'n' or continue_The_operations =='N':
        print("Thank you for using your Bank")
        break
    else:
        print('check your input error.... ')
        break
```
